```python
import matplotlib.pyplot as plt
import numpy as np
import csv
from matplotlib.pyplot import cm
plt.rcParams['text.usetex'] = True
%matplotlib inline
```

```python
##Import SMASH benchmarks

inflation_1_data=np.loadtxt('hc_inflation_SMASH_benchmark_1', dtype=float)
inflation_2_data=np.loadtxt('hc_inflation_SMASH_benchmark_2', dtype=float)
preheating_1_data=np.loadtxt('hc_preheating_SMASH_benchmark_1', dtype=float)
preheating_2_data=np.loadtxt('hc_preheating_SMASH_benchmark_2', dtype=float)
thermal_1_data=np.loadtxt('hc_thermal_SMASH_benchmark_1', dtype=float)
thermal_2_data=np.loadtxt('hc_thermal_SMASH_benchmark_2', dtype=float)
```

```python
## Import data

dataBAW = np.loadtxt('BAW.txt', delimiter=',', dtype=float)
dataLSDweak = np.loadtxt('LSDweak.txt', delimiter=',', dtype=float)
dataLSDstrong = np.loadtxt('LSDstrong.txt', delimiter=',', dtype=float)
dataStrongARCADE = np.loadtxt('StrongARCADE.txt', delimiter=',', dtype=float)
dataWeakARCADE = np.loadtxt('WeakARCADE.txt', delimiter=',', dtype=float)
dataIAXOSPD = np.loadtxt('IAXOSPD.txt', delimiter=',', dtype=float)
dataIAXOHET = np.loadtxt('IAXOHET.txt', delimiter=',', dtype=float)
dataIAXO = np.loadtxt('IAXO.txt', delimiter=',', dtype=float)
dataALPSII = np.loadtxt('ALPSII.txt', delimiter=',', dtype=float)
dataJURA = np.loadtxt('JURA.txt', delimiter=',', dtype=float)
dataOSQAR = np.loadtxt('OSQAR.txt', delimiter=',', dtype=float)
dataCAST = np.loadtxt('CAST.txt', delimiter=',', dtype=float)
dataHOL = np.loadtxt('HOL.txt', delimiter=',', dtype=float)
dataAkutsu = np.loadtxt('Akutsu.txt', delimiter=',', dtype=float)
dataMagnonLow = np.loadtxt('MagnonLow.txt', delimiter=',', dtype=float)
dataMagnonHigh = np.loadtxt('MagnonHigh.txt', delimiter=',', dtype=float)
dataEDGESstrong = np.loadtxt('EDGESstrong.txt', delimiter=',', dtype=float)
dataEDGESweak = np.loadtxt('EDGESweak.txt', delimiter=',', dtype=float)

data=[dataHOL, dataAkutsu, dataMagnonLow, dataMagnonHigh, dataBAW, dataLSDweak, dataLSDstrong, dataWeakARCADE, dataStrongARCADE, dataEDGESstrong, dataEDGESweak, dataIAXOSPD, dataIAXOHET, dataIAXO, dataALPSII, dataJURA, dataOSQAR, dataCAST]
```

```python
####Load interferometers sensitivity curves
####
####
####

######ET

file_ET = open("plis_ET.dat","r")

file_ET_list = file_ET.readlines()

ET_data = np.array([[float(val) for val in line.split()] for line in file_ET_list[14:]])


file_ET.close()


######LISA

file_LISA = open("plis_LISA.dat","r")

file_LISA_list = file_LISA.readlines()

LISA_data = np.array([[float(val) for val in line.split()] for line in file_LISA_list[14:]])


file_LISA.close()


######DECIGO

file_DECIGO = open("plis_DECIGO.dat","r")

file_DECIGO_list = file_DECIGO.readlines()

DECIGO_data = np.array([[float(val) for val in line.split()] for line in file_DECIGO_list[14:]])


file_DECIGO.close()


######CE

file_CE = open("plis_CE.dat","r")

file_CE_list = file_CE.readlines()

CE_data = np.array([[float(val) for val in line.split()] for line in file_CE_list[14:]])


file_CE.close()


######BBO

file_BBO = open("plis_BBO.dat","r")

file_BBO_list = file_BBO.readlines()

BBO_data = np.array([[float(val) for val in line.split()] for line in file_BBO_list[14:]])


file_BBO.close()
```

```python
# Create Figure (empty canvas)
fig = plt.figure(figsize=(10, 8), dpi=400)

# Hol, Akutsu, MagnonLow, MagnonHigh, BAW, LSDweak, LSDstrong, StrongARCADE, WeakARCADE, StrongEDGES, WeakEDGES, IAXOSPD

n = len(data)
color = ['darkred','darksalmon', 'maroon', 'maroon', 'red', 'blue','blue','magenta','magenta','darkcyan',
         'cadetblue', 'cadetblue', 'orangered', 'aqua','springgreen','goldenrod','olive','sienna','lightcoral']
opacity = [0.3,0.5,1.,1.,0.3,0.3,0.5,0.5,0.5,0.3,0.3,0.3,0.3,0.3,0.3,0.3,0.3,0.3,0.3,0.3]

# Add set of axes to figure
axes = fig.add_axes([1., 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

axes.set_xscale("log", base=10)
axes.set_yscale("log", base=10)
axes.set_xlim([fmin, fmax])
axes.set_ylim([hmin, hmax])




# Plot on that set of axes
i=0
for d in data:
    axes.plot(d[:,0], d[:,1], c=color[i], alpha=opacity[i])
    axes.set_xlabel(r'$f\,\,{\rm [Hz]}$', fontsize = 14) # Notice the use of set_ to begin methods
    axes.set_ylabel(r'$h_c$', fontsize = 14)
    axes.fill_between(d[:,0], d[:,1], y2 = hmax, color=color[i], alpha=opacity[i])
    i=i+1
    

#plot interferometers
#x,y coords from interferometers
x_LISA = 10**LISA_data[:,0]
y_LISA =  10**LISA_data[:,3]
x_ET = 10**ET_data[:,0]
y_ET = 10**ET_data[:,3]
x_DECIGO = 10**DECIGO_data[:,0]
y_DECIGO =  10**DECIGO_data[:,3]
x_CE = 10**CE_data[:,0]
y_CE = 10**CE_data[:,3]
x_BBO = 10**BBO_data[:,0]
y_BBO =  10**BBO_data[:,3]


axes.loglog(x_LISA,y_LISA,'#9c2e17ff')
axes.loglog(x_ET,y_ET,'#8fe3b0ff')
axes.loglog(x_DECIGO,y_DECIGO,'#abababff')
axes.loglog(x_CE,y_CE,'#e39c9cff')
axes.loglog(x_BBO,y_BBO,'#7de3ffff')

#plot SMASH benchmarks

x = inflation_1_data[:,0]
y1 = inflation_1_data[:,1]+preheating_1_data[:,1]+thermal_1_data[:,1]
y2 = inflation_2_data[:,1]+preheating_2_data[:,1]+thermal_2_data[:,1]
axes.loglog(x, y1, color='#fdbf84ff')
axes.loglog(x, y2, color='#fd7d00ff')


plt.fill_between(x, y1, y2, color='orange',
                 alpha=0.5)

##### Customs ticks and labels x axis

xlabels=['$10^{'+str(-3+val)+'}$' for val in range(24)]
xlabels[3]='1'

for i in range(12):
   xlabels[2*i] = ''


xticks1= [10.**float(-3+val) for val in range(24)]

xticks2=  [float(2+val2)*10.**float(-3+val) for val in range(24) for val2 in range(8)]


yticks= [10.**float(-40+val) for val in range(31)]


ylabels=['$10^{'+str(-40+val)+'}$' for val in range(31)]

for i in range(31):
   if i%5!=0:
       ylabels[i] = ''



axes.tick_params(which='both',axis='both',direction='in')

axes.set_xticks(xticks1,labels=xlabels)

axes.set_xticks(xticks2,minor=True)


axes.set_yticks(yticks,labels=ylabels)


def id(x):
    return x

secax = axes.secondary_xaxis('top', functions=(id,id))

secax2 =  axes.secondary_yaxis('right', functions=(id,id))



secax.tick_params(which='both',axis='both',direction='in')


secax2.tick_params(axis='both',direction='in')

secax.set_xticks(xticks1,labels=['' for val in range(24)])

secax.set_xticks(xticks2,minor=True)


secax2.set_yticks(yticks,labels=['' for val in range(31)])


axes.xaxis.labelpad = 10
axes.yaxis.labelpad = 10

#angle_data = np.rad2deg(np.arctan((np.log10(y1[-1,1])-np.log10(y1[1,1]))/(np.log10(y1[-1,0])-np.log10(y1[1,0]))))
#angle_data_2 = np.rad2deg(np.arctan((y1[-1,1]-y1[1,1])/(y1[-1,0]-y1[1,0])))
#angle_screen = axes.transData.transform_angles(np.array((angle_data_2,)), 
                                             # np.array([y1[-1,0], y1[-1,1]]).reshape((1, 2)))[0]
#angle_screen_2 = axes.transData.transform_angles(np.array((angle_data_2,)), 
                                              #np.array([y1[-1,0], y1[-1,1]]).reshape((1, 2)))[0]
#loc = np.array((y1[-1,0], y1[-1,1]))
#axes.text(*loc, r'\textbf{source}', color='b', rotation = 60, rotation_mode='anchor', transform_rotates_text=True)
#axes.annotate("Text", xy=(y1[-1,0],y1[-1,1]), xytext=(2,2), textcoords="offset points",
              #rotation_mode='anchor', rotation=angle_data_2)

axes.set_title(r'\textbf{UHF-GW plotter}', fontsize = 18, pad = 20)


```
